```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv('data/titanic/train.csv')
data_df.head()
```

# Bar Graph

```python
frequencies = data_df['Pclass'].value_counts()
print(frequencies)

data_df['Pclass'].value_counts().plot(kind='bar')
```

# Pie Chart

```python
data_df['Pclass'].value_counts().plot(kind='pie', autopct='%0.2f')
```

```python
freq_prc = ((frequencies/frequencies.sum())*100)
freq_prc
```

# Cumulative Freq

```python
c = []

cf = 0
for i in freq_prc:
    cf = cf+i
    c.append(cf)

print(c)

plt.plot(c, marker='d')
```
