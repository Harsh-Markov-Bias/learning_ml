```python
import pandas as pd
import numpy as np 
```

```python
import time
```

```python
import requests
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)',}
```

```python
df_list = []
not_done = []
for j in range(1,501):

    success = False
    retry=5

    while not success and retry>0:

        try:
            url = 'https://api.themoviedb.org/3/movie/popular?api_key=0db8bd1ca405744b00d98b4a9db4a56d&language=en-US&page={}'.format(j)

            response = requests.get(url, headers=HEADERS, timeout=10).json()

            df = pd.DataFrame(response['results'])
            df = df[['title', 'overview', 'original_language', 'release_date', 'vote_average', 'popularity', 'vote_count']]

            df_list.append(df)
            print(f'page {j} is done')
            success = True
        except Exception as e:
            print(f"⚠️ Error on page {j}, retries left {retry - 1}: {e}")
            retry -= 1
            time.sleep(2) 
    if not success:
        not_done.append(j)
        print(f"page {j} not possible after 5 retries")

    time.sleep(0.1)


final_df = pd.concat(df_list, ignore_index=True)

```
